- Mean
- Median
- Standard Deviation
- Variance
- Minimum
- Maximum
- Range
- Kurtosis : A measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution

```python
# bending1
bending1_features = []

bending1_path = '../data/bending1'

for dataset_num in range(1, 8):
    file_path = f"{bending1_path}/dataset{dataset_num}.csv"

    df = pd.read_csv(file_path, skiprows=4).drop('# Columns: time', axis=1)
    
    df_features = {
        'Instance': f'{dataset_num}'
    }
    for i, column in enumerate(df.columns, start=1):
        df_features[f'min_{i}'] = df[column].min()
        df_features[f'max_{i}'] = df[column].max()
        df_features[f'std_{i}'] = df[column].std()
        df_features[f'mean_{i}'] = df[column].mean()
        df_features[f'median_{i}'] = df[column].median()
        df_features[f'1st_quart_{i}'] = df[column].quantile(0.25)
        df_features[f'3rd_quart_{i}'] = df[column].quantile(0.75)
    
    bending1_features.append(df_features)
    
bending1_extract = pd.DataFrame(bending1_features)
columns = ['Instance'] + [col for col in bending1_extract if col != 'Instance']
bending1_extract = bending1_extract[columns]


# bending2
bending2_features = []

bending2_path = '../data/bending2'

for dataset_num in range(1, 7):
    file_path = f"{bending2_path}/dataset{dataset_num}.csv"

    df = pd.read_csv(file_path, skiprows=4).drop('# Columns: time', axis=1)
    
    df_features = {
        'Instance': f'{dataset_num}'
    }
    for i, column in enumerate(df.columns, start=1):
        df_features[f'min_{i}'] = df[column].min()
        df_features[f'max_{i}'] = df[column].max()
        df_features[f'std_{i}'] = df[column].std()
        df_features[f'mean_{i}'] = df[column].mean()
        df_features[f'median_{i}'] = df[column].median()
        df_features[f'1st_quart_{i}'] = df[column].quantile(0.25)
        df_features[f'3rd_quart_{i}'] = df[column].quantile(0.75)
    
    bending2_features.append(df_features)
    
bending2_extract = pd.DataFrame(bending2_features)
columns = ['Instance'] + [col for col in bending2_extract if col != 'Instance']
bending2_extract = bending2_extract[columns]


# cycling
cycling_features = []

for dataset_num in range(1, 16):
    file_path = f"{cycling_path}/dataset{dataset_num}.csv"

    df = pd.read_csv(file_path, skiprows=4).drop('# Columns: time', axis=1)
    
    df_features = {
        'Instance': f'{dataset_num}'
    }
    for i, column in enumerate(df.columns, start=1):
        df_features[f'min_{i}'] = df[column].min()
        df_features[f'max_{i}'] = df[column].max()
        df_features[f'std_{i}'] = df[column].std()
        df_features[f'mean_{i}'] = df[column].mean()
        df_features[f'median_{i}'] = df[column].median()
        df_features[f'1st_quart_{i}'] = df[column].quantile(0.25)
        df_features[f'3rd_quart_{i}'] = df[column].quantile(0.75)
    
    cycling_features.append(df_features)
    
cycling_extract = pd.DataFrame(cycling_features)
columns = ['Instance'] + [col for col in cycling_extract if col != 'Instance']
cycling_extract = cycling_extract[columns]

# lying
lying_features = []

for dataset_num in range(1, 16):
    file_path = f"{lying_path}/dataset{dataset_num}.csv"

    df = pd.read_csv(file_path, skiprows=4).drop('# Columns: time', axis=1)
    
    df_features = {
        'Instance': f'{dataset_num}'
    }
    for i, column in enumerate(df.columns, start=1):
        df_features[f'min_{i}'] = df[column].min()
        df_features[f'max_{i}'] = df[column].max()
        df_features[f'std_{i}'] = df[column].std()
        df_features[f'mean_{i}'] = df[column].mean()
        df_features[f'median_{i}'] = df[column].median()
        df_features[f'1st_quart_{i}'] = df[column].quantile(0.25)
        df_features[f'3rd_quart_{i}'] = df[column].quantile(0.75)
    
    lying_features.append(df_features)
    
lying_extract = pd.DataFrame(lying_features)
columns = ['Instance'] + [col for col in lying_extract if col != 'Instance']
lying_extract = lying_extract[columns]


# Sitting
sitting_features = []

for dataset_num in range(1, 16):
    file_path = f"{sitting_path}/dataset{dataset_num}.csv"

    df = pd.read_csv(file_path, skiprows=4).drop('# Columns: time', axis=1)
    
    df_features = {
        'Instance': f'{dataset_num}'
    }
    for i, column in enumerate(df.columns, start=1):
        df_features[f'min_{i}'] = df[column].min()
        df_features[f'max_{i}'] = df[column].max()
        df_features[f'std_{i}'] = df[column].std()
        df_features[f'mean_{i}'] = df[column].mean()
        df_features[f'median_{i}'] = df[column].median()
        df_features[f'1st_quart_{i}'] = df[column].quantile(0.25)
        df_features[f'3rd_quart_{i}'] = df[column].quantile(0.75)
    
    sitting_features.append(df_features)
    
sitting_extract = pd.DataFrame(sitting_features)
columns = ['Instance'] + [col for col in sitting_extract if col != 'Instance']
sitting_extract = sitting_extract[columns]

# standing
standing_features = []

for dataset_num in range(1, 16):
    file_path = f"{standing_path}/dataset{dataset_num}.csv"

    df = pd.read_csv(file_path, skiprows=4).drop('# Columns: time', axis=1)
    
    df_features = {
        'Instance': f'{dataset_num}'
    }
    for i, column in enumerate(df.columns, start=1):
        df_features[f'min_{i}'] = df[column].min()
        df_features[f'max_{i}'] = df[column].max()
        df_features[f'std_{i}'] = df[column].std()
        df_features[f'mean_{i}'] = df[column].mean()
        df_features[f'median_{i}'] = df[column].median()
        df_features[f'1st_quart_{i}'] = df[column].quantile(0.25)
        df_features[f'3rd_quart_{i}'] = df[column].quantile(0.75)
    
    standing_features.append(df_features)
    
standing_extract = pd.DataFrame(standing_features)
columns = ['Instance'] + [col for col in standing_extract if col != 'Instance']
standing_extract = standing_extract[columns]

# walking
walking_features = []

for dataset_num in range(1, 16):
    file_path = f"{walking_path}/dataset{dataset_num}.csv"

    df = pd.read_csv(file_path, skiprows=4).drop('# Columns: time', axis=1)
    
    df_features = {
        'Instance': f'{dataset_num}'
    }
    for i, column in enumerate(df.columns, start=1):
        df_features[f'min_{i}'] = df[column].min()
        df_features[f'max_{i}'] = df[column].max()
        df_features[f'std_{i}'] = df[column].std()
        df_features[f'mean_{i}'] = df[column].mean()
        df_features[f'median_{i}'] = df[column].median()
        df_features[f'1st_quart_{i}'] = df[column].quantile(0.25)
        df_features[f'3rd_quart_{i}'] = df[column].quantile(0.75)
    
    walking_features.append(df_features)
    
walking_extract = pd.DataFrame(walking_features)
columns = ['Instance'] + [col for col in walking_extract if col != 'Instance']
walking_extract = walking_extract[columns]
```

```python
def bootstrap_std(data, n_bootstrap=1000, ci=90):
    fill_value = np.nanmean(data)
    data_filled = np.where(np.isnan(data), fill_value, data)
    
    bootstrapped_stds = []
    
    for _ in range(n_bootstrap):
        sample = np.random.choice(data_filled, size=len(data_filled),replace = True)
        std = np.std(sample, ddof = 1)
        bootstrapped_stds.append(std)
        
    lower_percentile = (100 - ci)/2
    upper_percentile = 100 - lower_percentile
    confidence_interval = np.percentile(bootstrapped_stds, [lower_percentile, upper_percentile])
    return np.mean(bootstrapped_stds), confidence_interval


bootstrap_results = {}
for feature in combined_df.columns:
    mean_std, ci = bootstrap_std(combined_df[feature].values)
    bootstrap_results[feature] = {'Mean Std': mean_std, '90% CI': ci}
    
data_for_df = [{
    'Feature': feature,
    'Mean Std Dev': stats['Mean Std'],
    'CI Lower Bound': stats['90% CI'][0],
    'CI Upper Bound': stats['90% CI'][1]
} for feature, stats in bootstrap_results.items()]


bootstrap_df = pd.DataFrame(data_for_df)

bootstrap_df.set_index('Feature', inplace=True)
bootstrap_df
```

## 2. ISLR 3.7.4

### (a) Linear Train

If the true relationship between X and Y is linear, then we would expect that simple linear regression model fit the data better than the cubic regression. Cubic regression model would not improve the fir because the features they learned might not be the true relationship between X and Y. Since the cubic regression is more complex thus the cubic regression would have lower training RSS than the simple linear regression which might be overfitting.

### (b) Linear Test

The test data is something that the training model has never seen, thus the cubic regression model would learn training data very well but not on the test data. In other words, the simple linear regression would have a lower test RSS than the cubic regression because the cubic regression would have overfitting features from the training dataset. And since the relationship between X and Y is linear so the simple linear regression would perform better with the test RSS

### (c) Not Linear Train

If the relationship between X and Y is not linear, then the linear regression model might not having a good prediction because it is not linear relationship. The cubic model which is more complex statistical model would learn the non-linear relationship much better, and easily catch the hidden features from the data. Thus, we will expect the cubic regression would have lower training RSS than the linear regression model

### (d) Not Linear Testing

Since the relationship is not linear, and the cubic model fits better in the train RSS. Thus, the cubic regression model should still be better than the linear regression in the test dataset. However, if the training data contains too many random noise then the cubic regression model might do worse on the test data. But we still expect the cubic regression model do the better prediction than the simple linear regression.
